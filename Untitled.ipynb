```python
WIDTH = 1920
HEIGHT = 1080

def get_background_image(img):
    h_ratio = HEIGHT / img.shape[0]
    w_ratio = WIDTH / img.shape[1]
    bgd = cv2.resize(img, (0, 0), fx=w_ratio, fy=h_ratio)
    bgd = cv2.blur(bgd, (HEIGHT//15,  WIDTH//15))
    bgd = (.8 * bgd).astype('uint8')
    return bgd

def resize_img(img):
    h_ratio = img.shape[0] / HEIGHT
    w_ratio = img.shape[1] / WIDTH
    ratio = max(h_ratio, w_ratio)
    ratio = .9/ratio if ratio > 1 else 1
    img = cv2.resize(img, (0, 0), fx=ratio, fy=ratio)
    return img

def prettify(img):
    bgd = get_background_image(img)
    img = resize_img(img)
    img_h, img_w, _ = img.shape
    off_h = (HEIGHT - img_h) // 2
    off_w = (WIDTH - img_w) // 2
    bgd[off_h:off_h+img_h, off_w:off_w+img_w, :] =img
    return bgd

def annotate(img, title):
    font                   = cv2.FONT_HERSHEY_COMPLEX
    bottomLeftCornerOfText = (10, HEIGHT - 10)
    fontScale              = 1
    fontColor              = (255,255,255)
    lineType               = 2
    cv2.putText(img, title, 
        bottomLeftCornerOfText, 
        font, 
        fontScale,
        fontColor,
        lineType)
    return img

for file in glob.glob('/home/pdesahb/Pictures/background/raw/*'):
    filename = file.split('/')[-1]
    title = '.'.join(file.split('/')[-1].split('.')[:-1])
    img = cv2.imread(file, cv2.IMREAD_COLOR)
    new_img = annotate(prettify(img), title)
    cv2.imwrite('/home/pdesahb/Pictures/background/' + filename, new_img)
```
